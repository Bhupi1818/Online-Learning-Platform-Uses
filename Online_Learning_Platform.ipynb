```python
# Average Hours Per Week on Online Platform by Top 10 Users


top_10_user = df.groupby('user_id')['hours_per_week'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.plot(top_10_user.index,top_10_user.values,marker='o')

plt.grid(color='red',linestyle=':')
plt.title('Average Hours Per Week on Online Platform by Top 10 Users')
plt.xlabel('User ID')
plt.ylabel('Hours Per Week')

plt.tight_layout()
plt.show()
```

```python
# Average Completion Rate Per Platform

avg_completion = df.groupby('primary_platform')['completion_rate_percent'].mean()

plt.figure(figsize=(10,6))

plt.plot(avg_completion.index, avg_completion.values, marker='o')

plt.grid(color='red',linestyle=':')
plt.title('Average Completion Rate Per Platform')
plt.xlabel('Platform')
plt.ylabel('Completion Rate (%)')

plt.show()
```

```python
# Average Timing Per Platform

avg_time = df.groupby('primary_platform')['hours_per_week'].mean()

plt.plot(avg_time.index,avg_time.values,marker='o')

plt.grid(color='red',linestyle=':')
plt.title('Average Timing Per Platform ')
plt.xlabel('Platform')
plt.ylabel('Time (Hours) ')
plt.xticks(rotation=45)
plt.show()
```

```python
# Learning Devices Used


device = df['learning_device'].value_counts().head()

plt.figure(figsize=(10,8))

plt.pie(device.values,labels=device.index,autopct='%1.1f%%')

plt.title('Learning Devices Used')

plt.show()
```

```python
# Learning Purpose


purpose = df['learning_purpose'].value_counts().head()

plt.figure(figsize=(10,6))

plt.pie(purpose.values,labels=purpose.index,autopct='%1.1f%%')

plt.title('Learning Purpose')

plt.tight_layout()

plt.show()
```

```python
## Content Type Preference


content = df['content_type_preference'].value_counts().head()

plt.figure(figsize=(10,6))

plt.pie(content.values,labels=content.index,autopct='%1.1f%%')

plt.title('Most Prefered Content')
plt.tight_layout()
plt.show()
```

```python
#Age Group 

group = df['age_group'].value_counts().head()

plt.figure(figsize=(10,6))

plt.pie(group.values, labels=group.index, autopct='%1.1f%%')

plt.title('Age group Percentage Which use online learning platform')
plt.tight_layout()
plt.show()
```

```python
## Role For Which Platform is Used

role = df['role'].value_counts().head()

plt.figure(figsize=(10,6))

plt.pie(role.values, labels=role.index, autopct='%1.1f%%')

plt.title('Role For Which Platform is Used')
plt.tight_layout()
plt.show()
```

```python
#Primary Platform used


platform = df['primary_platform'].value_counts().head()

plt.figure(figsize=(10,6))

plt.pie(platform.values, labels=platform.index, autopct='%1.1f%%')

plt.title('Primary Platforms Used')
plt.tight_layout()
plt.show()
```

```python
# Age_Group VS Learning_Purpose

score = df.groupby('primary_platform')['satisfaction_score'].mean()

plt.figure(figsize=(10,6))

plt.bar(score.index,score.values)

plt.title('Satisfaction Score Per Platform ')
plt.xlabel('Platform')
plt.ylabel('Satisfaction Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Age_Group VS Learning_Purpose

plt.figure(figsize=(10,8))

pivot = pd.crosstab(df['age_group'],df['learning_purpose'])


pivot.plot(kind='bar')

plt.title('Age_Group VS Learning_Purpose')
plt.ylabel('Count')
plt.xlabel('Age Group')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# Age VS Role

hours = df.groupby('role')['hours_per_week'].mean()

plt.figure(figsize=(8,6))

plt.bar(hours.index,hours.values)

plt.title('How Many Hours Of Work done by a Role Per Week')
plt.xlabel('Role')
plt.ylabel('Hours Per Week')
plt.xticks(rotation=45)
plt.show()
```
